```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('birds.csv')

# Select row 4 (index 3, since Python is 0-based)
snow_goose = df.iloc[3]

# Prepare data for plotting
attributes = ['MinLength', 'MaxLength', 'MinBodyMass', 'MaxBodyMass', 'MinWingspan', 'MaxWingspan']
values = [snow_goose[attr] for attr in attributes]

# 1. Line plot
plt.figure(figsize=(8,5))
plt.plot(attributes, values, marker='o', linestyle='-', color='b')
plt.title('Snow Goose Measurements')
plt.ylabel('Value')
plt.xlabel('Attribute')
plt.tight_layout()
plt.show() # Changed from plt.savefig()
plt.close()

# 2. Scatter plot
plt.figure(figsize=(8,5))
plt.scatter(attributes, values, color='r')
plt.title('Snow Goose Measurements (Scatter)')
plt.ylabel('Value')
plt.xlabel('Attribute')
plt.tight_layout()
plt.show() # Changed from plt.savefig()
plt.close()

# 3. Bar plot
plt.figure(figsize=(8,5))
plt.bar(attributes, values, color='g')
plt.title('Snow Goose Measurements (Bar)')
plt.ylabel('Value')
plt.xlabel('Attribute')
plt.tight_layout()
plt.show() # Changed from plt.savefig()
plt.close()
```
